```python
# Install Kaggle API
!pip install kaggle

# Upload kaggle.json (you need to upload the file manually to Colab)
import os
os.makedirs('/root/.kaggle', exist_ok=True)
!mv kaggle.json /root/.kaggle/
!chmod 600 /root/.kaggle/kaggle.json

# Download the Titanic dataset
!kaggle competitions download -c titanic -f train.csv
!unzip train.csv.zip

# Load and clean the dataset
import pandas as pd

# Load the dataset
df = pd.read_csv('train.csv')
print("Initial DataFrame:")
print(df.head())

# Check for missing values
print("\nMissing values before cleaning:")
print(df.isnull().sum())

# Fill missing values for 'Age' with the median
df['Age'].fillna(df['Age'].median(), inplace=True)

# Fill missing values for 'Embarked' with the mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Drop the 'Cabin' column due to a high number of missing values
df.drop(columns=['Cabin'], inplace=True)

# Convert 'Sex' to numerical values
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})

# Convert 'Embarked' to numerical values
df['Embarked'] = df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})

# Drop columns that are not useful for prediction
df.drop(columns=['Name', 'Ticket', 'PassengerId'], inplace=True)

print("\nCleaned DataFrame:")
print(df.head())

print("\nMissing values after cleaning:")
print(df.isnull().sum())

# Save the cleaned DataFrame to a new CSV file
df.to_csv('analyzed_train.csv', index=False)
print("\nCleaned DataFrame has been saved to 'analyzed_train.csv'")

# Initial analysis
print("\nSummary Statistics:")
print(df.describe())

import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of Age
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.savefig('age_distribution.png')  # Save the plot to a file
plt.show()

# Survival rate by sex
plt.figure(figsize=(10, 6))
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Sex')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.savefig('survival_rate_by_sex.png')  # Save the plot to a file
plt.show()

# Survival rate by Embarked
plt.figure(figsize=(10, 6))
sns.barplot(x='Embarked', y='Survived', data=df)
plt.title('Survival Rate by Embarked')
plt.xlabel('Embarked')
plt.ylabel('Survival Rate')
plt.savefig('survival_rate_by_embarked.png')  # Save the plot to a file
plt.show()
```
